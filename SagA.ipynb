## Sagatarius A* $\theta_E$ distribution

$M_{SagA*} = 4.0\times10^6\,M_\odot$, $D_L=D_{SagA*}=8.27\,\rm{kpc}$ 

```python
import numpy as np
import matplotlib.pyplot as plt

def theta_E(M,DL,DS):
    ''' Find theta E from M, DL, and DS
    
    inputs:
    --------
    M:          in solar masses
    DL:         in pc
    DS:         in pc
    
    outputs:
    ---------
    theta_E:    in mas'''

    pi_rel = (1.0/DL - 1.0/DS)/1000.  # mas
    kappa = 8.144  # mas/solar mass

    return np.sqrt(M*kappa*pi_rel)

def theta_E_SagA(DS):
    ''' Find the distribution of possible theta E for Sgatarius A*
    
    inputs:
    --------
    DS:         in pc (must be > 8270)
    
    outputs:
    ---------
    theta_E:    in mas'''

    D_L = 8270  # pc
    M = 4.0 *10**(6.0)  # solar masses

    return theta_E(M, D_L, DS)

D_S = np.linspace(8280, 16500, 822)
theta_E = theta_E_SagA(D_S)

plt.figure(dpi=600)
plt.plot(D_S/1000.,theta_E*1000.,'k-')
plt.ylabel(r'$\theta_E$ ($\mu$as)')
plt.xlabel(r'Source distance, $D_S\, (\rm kpc)$')
plt.savefig('SagA_ThetaE.png')
plt.show()
```
