## Sagatarius A* $\theta_E$ distribution

$M_{SagA*} = 4.0\times10^6\,M_\odot$, $D_L=D_{SagA*}=8.27\,\rm{kpc}$ 

```python
import numpy as np
import matplotlib.pyplot as plt

def get_theta_E(M,DL,DS):
    ''' Find theta E from M, DL, and DS
    
    inputs:
    --------
    M:          in solar masses
    DL:         in pc
    DS:         in pc
    
    outputs:
    ---------
    theta_E:    in mas'''

    pi_rel = (1.0/DL - 1.0/DS)*1000.  # mas
    kappa = 8.144  # mas/solar mass

    return np.sqrt(M*kappa*pi_rel)

def theta_E_SagA(DS):
    ''' Find the distribution of possible theta E for Sgatarius A*
    
    inputs:
    --------
    DS:         in pc (must be > 8270)
    
    outputs:
    ---------
    theta_E:    in mas'''

    D_L = 8270  # pc
    M = 4.30 *10**(6.0)  # solar masses

    return get_theta_E(M, D_L, DS)

D_S = np.linspace(8270.00001, 8270.7, 500)
theta_E = theta_E_SagA(D_S)

plt.figure(dpi=600)
plt.plot(D_S/1000.,theta_E,'k:', alpha=0.7, label=r'SagittariusA*: $M=4\times10^6\,M_\odot$ and $D_L=8.3\,\rm{kpc}$')

D_S = np.linspace(8271, 9000, 500)
theta_E = get_theta_E(3., 8270., D_S)
plt.plot(D_S/1000.,theta_E,'k-', label=r'$M=3\,M_\odot$, $D_L=8.3\,\rm{kpc}$')
theta_E = get_theta_E(100., 8270., D_S)
plt.plot(D_S/1000.,theta_E,'k:', label=r'$M=100\,M_\odot$')

D_S = np.linspace(6001, 9000, 500)
theta_E = get_theta_E(3., 6000., D_S)
plt.plot(D_S/1000.,theta_E,'m-', label=r'$M=3\,M_\odot$, $D_L=6\,\rm{kpc}$')
theta_E = get_theta_E(100., 6000., D_S)
plt.plot(D_S/1000.,theta_E,'m:', label=r'$M=100\,M_\odot$ ')

D_S = np.linspace(4001, 9000, 500)
theta_E = get_theta_E(3., 4000, D_S)
plt.plot(D_S/1000.,theta_E,'y-', label=r'$M=3\,M_\odot$, $D_L=4\,\rm{kpc}$')
theta_E = get_theta_E(100., 4000., D_S)
plt.plot(D_S/1000.,theta_E,'y:', label=r'$M=100\,M_\odot$')

D_S = np.linspace(2001, 9000, 500)
theta_E = get_theta_E(3., 2000., D_S)
plt.plot(D_S/1000.,theta_E,'c-', label=r'$M=3\,M_\odot$, $D_L=2\,\rm{kpc}$')
theta_E = get_theta_E(100., 2000., D_S)
plt.plot(D_S/1000.,theta_E,'c:', label=r'$M=100\,M_\odot$')

plt.axhline(0.009,color='grey', linestyle='dotted', linewidth=1, label='Einstein desert')
plt.axhline(0.025,color='grey', linestyle='dotted', linewidth=1)
plt.ylabel(r'$\theta_E$ (mas)')
plt.xlabel(r'Source distance, $D_S\, (\rm kpc)$')
plt.xlim(4,9)
plt.legend()
plt.savefig('BH_ThetaE.png')
plt.show()
```
